##### üçÅ amplitude

conceito semelhante ao escopo, por√©m entregamos apenas o reultado do maior menos o menor, ou seja:

o tamanho do escopo sem explicar onde come√ßa ou termina

ex.:

```python
from random import randint
dataset = [randint(1,100) for i in range(10)]
amplitude = max(dataset) - min(dataset)
print(f'dataset {dataset}')
print(f'escopo de {min(dataset)} √° {max(dataset)}')
print(f'amplitude √© de {amplitude}')
```

```python
from random import randint

def vari√¢ncia(dataset):
    from numpy import mean
    print(dataset)
    somat√≥rio = 0
    for i in range(len(dataset)):
        somat√≥rio += (i - mean(dataset))**2

    vari√¢ncia = round((somat√≥rio / len(dataset)), 3)
    return vari√¢ncia

vari√¢ncia([randint(1,20) for i in range(5)])
```
